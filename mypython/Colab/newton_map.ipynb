```python
from PIL import Image, ImageDraw
```

```python
def newton(x):
  for _ in range(10):
    # 以下を変更
    x = x - (x**4 - 1)/(4 * x **3)
  return x
```

```python
def plot(draw, s):
  # hsは真ん中から外ブチまでの距離
  hs = s //2
  red = (255,0,0)
  green = (0,255,0)
  blue = (0,0,255)
  purple = (255,0,255)
  for x in range(s):
    for y in range(s):
      # ピクセルの座標の正規化
      z = complex(x-hs+0.5, -y+hs+0.5) / s * 4
      z = newton(z)
      # 以下のブロックを変更
      if z.real + z.imag >= 0:
        if z.real - z.imag >= 0:
          c = red
        else:
          c = green
      else:
        if z.real - z.imag >= 0:
          c = blue
        else:
          c = purple
      draw.rectangle([x,y,x+1,y+1],fill=c)
```

```python
size = 512
img = Image.new("RGB",(size,size))
draw = ImageDraw.Draw(img)
# 書き込み
plot(draw,size)
# 表示
img
```
