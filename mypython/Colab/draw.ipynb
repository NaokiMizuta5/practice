```python
from PIL import Image, ImageDraw
from math import pi, sin , cos
```

```python
img = Image.new("L", (256,256), 255)
draw = ImageDraw.Draw(img)
cx = 128
cy = 128
r = 96
draw.ellipse((cx - r, cy -r, cx + r, cy + r))

N = 6
s = 2 * pi / N


# Nの偶奇で分岐
if N % 2 == 1:
  k = N // 2
  for i in range(N):
    s1 = ((i * k) % N) * s - 0.5 * pi
    s2 = s1 + s * k
    x1 = r * cos(s1) + cx
    y1 = r * sin(s1) + cy
    x2 = r * cos(s2) + cx
    y2 = r * sin(s2) + cy
    draw.line((x1,y1,x2,y2))
elif N == 6:
  k = 2
  for i in range(N):
    s1 = ((i * k) % N) * s - 0.5 * pi
    s2 = s1 + s * k
    x1 = r * cos(s1) + cx
    x11 = r * cos(s1+s) + cx
    y1 = r * sin(s1) + cy
    y11 = r * sin(s1+s) + cy
    x2 = r * cos(s2) + cx
    x21 = r * cos(s2+s) + cx
    y2 = r * sin(s2) + cy
    y21 = r * sin(s2+s) + cy
    draw.line((x1,y1,x2,y2))
    draw.line((x11,y11,x21,y21))
```
