```python
from graphviz import Digraph
from PIL import Image
```

```python
def collatz_graph(i, edges):
  while(i != 1 and i != 3):
    j = i
    if(i % 2 == 0):
      i = i // 2
    else:
      # i = i * 3 +1
      i = i * 3 +3
    edges.add((j,i))
```

```python
def make_graph(n):
  g = Digraph(format='png')
  edges = set()
  for i in range(1, n+1):
    collatz_graph(i,edges)
  for i, j in edges:
    g.edge(str(i), str(j))
  g.attr(size="20,20")
  g.render("test")
  return Image.open("test.png")
```

```python
make_graph(50)
```
